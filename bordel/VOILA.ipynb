```python
import numpy as np
from SecondOrderElec import LP, plot
import matplotlib.pyplot as plt
```

```python
for T,m in [(0.5, 0.1),(1, 0.5),(1.5, 1.5)]:
    T = LP(T0=T, m=m, w0=100)
    T.freqresp()
```

```python
r = np.linspace(0,4,1000)
for T,m in [(1, 0.1),(1, 0.5),(1, 1.5)]:
    T = LP(T0=T, m=m, w0=10)
    t,s = T.step(plot=False, T=r)
    plt.plot(t,s, label=f"T={T} et m={m}")
```
